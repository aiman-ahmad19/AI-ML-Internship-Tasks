# Task 1: Exploring and Visualizing a Simple Dataset

## Problem Statement and Goal
The goal of this task is to load, inspect, summarize, and visualize the Iris dataset using `pandas`, `matplotlib`, and `seaborn`. The analysis focuses on understanding data structure, trends, distributions, and potential outliers across flower species.


## Dataset Loading and Preprocessing
The dataset is stored locally as `iris.csv` and loaded with `pandas.read_csv()`. Basic preprocessing includes checking missing values and duplicate rows before analysis.


```python
import io

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
df = pd.read_csv('iris.csv')
print('Dataset loaded successfully.')
df.head()
```

```python
print('Shape:', df.shape)
print('Columns:', list(df.columns))
print('\nFirst five rows:')
print(df.head())
```

```python
print('Missing values by column:')
print(df.isnull().sum())
print('\nDuplicate rows:', df.duplicated().sum())
```

```python
buffer = io.StringIO()
df.info(buf=buffer)
print('Dataset info:')
print(buffer.getvalue())
```

```python
print('Summary statistics:')
df.describe(include="all")
```

## Data Visualization and Exploration
The visualizations below help reveal relationships between variables, distribution patterns, and possible outliers in the dataset.


```python
# Scatter plot showing the relationship between sepal length and petal length.
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', s=80)
plt.title('Sepal Length vs Petal Length by Species')
plt.show()
```

```python
# Histograms summarize the distribution of each numeric feature.
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df[numeric_columns].hist(bins=15, figsize=(12, 8), edgecolor='black')
plt.suptitle('Feature Distributions', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Box plots are useful for identifying spread and possible outliers.
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, column in zip(axes.flat, numeric_columns):
    sns.boxplot(data=df, x='species', y=column, ax=ax)
    ax.set_title(f"{column} by Species")
    ax.tick_params(axis="x", rotation=15)
plt.tight_layout()
plt.show()
```
